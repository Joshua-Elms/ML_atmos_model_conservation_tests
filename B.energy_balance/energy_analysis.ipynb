```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Load data

path_gen = lambda model: Path(f"/N/slate/jmelms/projects/ML_atmos_model_conservation_tests/B.energy_balance/data/nt=12_nic=1_NO_GCO/{model}_output.nc")
models = ["FuXi", "Pangu6x", "SFNO"]
load_model = "SFNO"
model_path = path_gen(load_model)
tmp_plot_dir = Path("./energy_plots")
tmp_plot_dir.mkdir(parents=True, exist_ok=True)
ds = xr.open_dataset(model_path)
```

```python
for t in ds.lead_time:
    plt.imshow(ds["total_energy"].sel(lead_time=t).squeeze())
    plt.title(f"Total Energy at lead time {t.item()}")
    plt.colorbar(label="Total Energy (J/m^2)")
    plt.savefig(tmp_plot_dir/f"total_energy_lead_time_{t.item()}.png")
    plt.clf()
    plt.close()
```
